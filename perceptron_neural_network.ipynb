```python
#Command used to maintain code reproducibility when working with random functions

import random
random.seed(1)
```

```python
import numpy as np
```

```python
#Create the input and output matrices. In this case, it is the truth table for the AND logical function

input = np.array([ [0,0], [0,1], [1,0], [1,1] ])
output = np.array([ [0,0,0,1] ]).T
```

```python
#Create the weight matrices for neural network - W

# The function random will create aleatory numbers between minimum and maximum pre determined

w1 = 2*np.random.random((2,4)) - 1
w2 = 2*np.random.random((4,1)) - 1
```

```python
w1
```

```python
w2
```

```python
#Neural network training

for interaction in range(1000000):
    layer1 = 1 / (1 + np.exp(-(np.dot(input, w1))))
    layer2 = 1 / (1 + np.exp(-(np.dot(layer1, w2))))
    layer2_delta = (output - layer2)*(layer2*(1-layer2))
    layer1_delta = layer2_delta.dot(w2.T) * (layer1 * (1-layer1))
    w2 += layer1.T.dot(layer2_delta)
    w1 += input.T.dot(layer1_delta)
```
