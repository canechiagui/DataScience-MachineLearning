    Trend Analysis: Moving Average Simulation
</h1>

<p> This notebook demonstrates a price simulation and the calculation of a Simple Moving Average (SMA)over a 5-period window. </p>  
<p> The goal is to show how to smooth out volatility in time-series data to better identify underlying market trends. </p>

```python
import numpy as np
```

```python
#Calculate moving average price

def moving_average(my_array, n = 5):
    values = np.cumsum(my_array, dtype = float)
    values[n:] = values[n:] - values[:-n]
    return values[n - 1:]/ n
```

```python
prices = np.random.uniform(10,10000, size = 100).round(2)

prices
```

```python
prices.sort()
prices
```

```python
p = 9
```

```python
mm = moving_average(prices, n = p).round(2)

mm
```
